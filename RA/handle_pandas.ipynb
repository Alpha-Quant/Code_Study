```python
with open("filter.csv", mode="w", encoding="utf-8") as fp_w:
    with open("output_tweets_utf8.csv", mode="r", encoding="utf-8") as fp_r:
        fp_r.readline()
        fp_w.write("date;retweets;favorites;text;id;link\n")  # 7항목
        lines = fp_r.readlines()
        for i, line in enumerate(lines):
            s = line.split(";")
            a = ";".join([s[1], s[2], s[3], s[4].strip("\""), s[8].strip("\""), s[9].strip().strip("\"")]) + "\n"
            if i<10:
                print(a)
            fp_w.write(a)
```

```python
filtered_df = df.drop_duplicates(subset="text")
#force_ascii 
```

```python
final_df = filtered_df[~filtered_df.isnull().T.any()]
```

```python
# final_df.to_json("tweet_result.json", orient="index",force_ascii=False)
```

```python
dic = final_df.to_dict(orient="index")
```

```python
from collections import OrderedDict
```

```python
dic2 = OrderedDict()
for key in dic.keys():
    dic2[str(key)] = dic[key]
```

```python
dic2
```

```python
import json
with open("tweet_result.json", mode="w", encoding="utf-8") as fp:
    json.dump(
        dic2,
        fp,
        ensure_ascii=False,
        sort_keys=True,
        indent=4,
        separators=(',', ': ')
    )
```
